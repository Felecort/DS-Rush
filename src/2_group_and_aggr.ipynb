```python
import pandas as pd
import numpy as np
```

```python
students_performance = pd.read_csv(
    "https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv")
students_performance.head()
```

```python
students_performance = students_performance.rename(columns={
    "parental level of education": "parental_level_of_education",
    "test preparation course": "test_preparation_course",
    "math score": "math_score",
    "reading score": "reading_score",
    "writing score": "writing_score",
}
)
```

```python

students_performance.groupby("gender").mean()
```

```python
students_performance.groupby("gender", as_index=0).aggregate({"math_score": "var",
                                                              "writing_score": "mean"})
```

```python
students_performance.groupby("gender").aggregate({"math_score": ["mean", "var"],
                                                  "writing_score": ["mean"]})
```

```python
students_performance.groupby("gender", as_index=0).aggregate({"math_score": ["mean", "var"],
                                                              "writing_score": ["mean"]})
```

```python
# multi indexes
students_performance.groupby(["gender", "race/ethnicity"]).aggregate({"math_score": ["mean", "var"],
                                                                      "writing_score": ["mean"]})
```

```python
students_performance.sort_values(
    ["gender", "math_score"], ascending=False).groupby("gender").head()
```

```python
students_performance["total_score"] = students_performance.math_score + \
    students_performance.reading_score + students_performance.writing_score
students_performance["total_score"]
```

```python
students_performance = students_performance.assign(
    totalscore_log=np.log(students_performance.total_score))
students_performance
```

```python
students_performance.drop(["total_score", "lunch"], axis=1)
```

```python
concentrations = pd.read_csv("http://stepik.org/media/attachments/course/4852/algae.csv")
# mean_concentrations = concentrations.groupby("genus").aggregate({"genus": "mean"})
concentrations.head()
```

```python

mean_concentrations = concentrations.groupby(["genus"]).mean()
mean_concentrations
```

```python
concentrations_fucus = concentrations.filter(items=["genus", "alanin"])
concentrations_fucus = concentrations_fucus.where(concentrations_fucus.genus=="Fucus")
concentrations_fucus.describe()

```

```python
concentrations.groupby('genus').agg(['min', 'mean', 'max'])
```

```python
concentrations.query('genus == "Fucus"').alanin.describe()
```

```python
concentrations.groupby("group").aggregate({"sucrose": ["std", "count"]})
```

```python
concentrations.groupby('group') \
    .agg({'citrate': 'var', 'sucrose': lambda x: x.max() - x.min(), 'glucose': 'count'}) \
    .round(2)
```
